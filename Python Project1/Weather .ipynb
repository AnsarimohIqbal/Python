# Detect OutLiers 

```python
plt.figure(figsize=(8,4))
plt.boxplot(df["Temperature"] , patch_artist=True , boxprops=dict(facecolor="skyblue"),
    medianprops=dict(color="red", linewidth=2),
    whiskerprops=dict(color="gray"),
    capprops=dict(color="gray"),
    flierprops=dict(marker='o', markerfacecolor='orange', markersize=6))
plt.title("Distribution and Outliers of Temperature", fontsize=14)
plt.ylabel("Temperature")
plt.show()
```

# Season Wise Average Temperature 

```python
Average_Temperature = df.groupby("Season")["Temperature"].mean().reset_index(name="Avg Temperature")
```

```python
Average_Temperature
```

```python
plt.figure(figsize=(5,4))

sns.barplot( x="Season",y="Avg Temperature",data=Average_Temperature, palette="Set2")

plt.title("Season-wise Average Temperature")
plt.xlabel("Season")
plt.ylabel("Average Temperature")
plt.show()
```

# Season Wise Wind Speed

```python
season_wind = df.groupby("Season")["Wind Speed"].mean().reset_index()
```

```python
season_wind
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x="Season", y="Wind Speed", data=season_wind, palette="Set3")

plt.title("Season-wise Average Wind Speed")
plt.xlabel("Season")
plt.ylabel("Average Wind Speed")
plt.show()
```

# Location Wise Weather Type (Categorical Vs Categorical)

```python
plt.figure(figsize=(10,6))

sns.countplot(x="Location", hue="Weather Type", data=df, palette="Set2")

plt.title("Location-wise Distribution of Weather Types", fontsize=14)
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Weather Type")
plt.show()
```

# Location Wise Temperature 

```python
plt.figure(figsize=(10,8))

sns.boxplot( x="Location", y="Temperature", data=df, palette="Set3")

plt.title("Temperature Distribution by Location", fontsize=14)
plt.xlabel("Location")
plt.ylabel("Temperature")
plt.xticks(rotation=45)
plt.show()
```

# Location Wise Wind Speed 

```python
plt.figure(figsize=(10,6))

sns.boxplot( x="Location", y="Wind Speed", data=df, palette="Set2")

plt.title("Wind Speed Distribution by Location", fontsize=14)
plt.xlabel("Location")
plt.ylabel("Wind Speed")
plt.xticks(rotation=45)
plt.show()
```
