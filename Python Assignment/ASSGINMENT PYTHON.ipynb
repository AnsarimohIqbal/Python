# School Management System

```python
class SchoolManagement:
    def __init__(self):
        self.students = {}
        self.student_id = 1000

    def new_admission(self):
        name = input("Student Name: ")
        age = int(input("Age: "))
        if age < 5 or age > 18:
            print(" Invalid age")
            return

        student_class = int(input("Class (1-12): "))
        mobile = input("Guardian Mobile: ")

        if len(mobile) != 10 or not mobile.isdigit():
            print(" Invalid mobile number")
            return

        self.student_id += 1
        self.students[self.student_id] = {
            "name": name,
            "age": age,
            "class": student_class,
            "mobile": mobile
        }

        print(f" Admission Successful. Student ID: {self.student_id}")

    def view_student(self):
        sid = int(input("Enter Student ID: "))
        print(self.students.get(sid, " Student not found"))

    def update_student(self):
        sid = int(input("Student ID: "))
        if sid not in self.students:
            print(" Student not found")
            return

        print("1. Update Mobile")
        print("2. Update Class")
        choice = input("Choose: ")

        if choice == "1":
            mobile = input("New Mobile: ")
            self.students[sid]["mobile"] = mobile
        elif choice == "2":
            self.students[sid]["class"] = int(input("New Class: "))

        print(" Updated successfully")

    def remove_student(self):
        sid = int(input("Student ID: "))
        if self.students.pop(sid, None):
            print("❌ Student removed")
        else:
            print("Student not found")

    def menu(self):
        while True:
            print("\n--- School Management ---")
            print("1. New Admission")
            print("2. View Student")
            print("3. Update Student")
            print("4. Remove Student")
            print("5. Exit")

            choice = input("Enter choice: ")

            if choice == "1":
                self.new_admission()
            elif choice == "2":
                self.view_student()
            elif choice == "3":
                self.update_student()
            elif choice == "4":
                self.remove_student()
            elif choice == "5":
                break
```

# Bus Reservation System

```python
class BusReservation:
    def __init__(self):
        self.routes = {
            "Mumbai-Pune": 500,
            "Delhi-Jaipur": 600,
            "Ahmedabad-Surat": 400
        }
        self.tickets = {}
        self.ticket_id = 2000
        self.seats = {route: 0 for route in self.routes}

    def show_routes(self):
        for route, price in self.routes.items():
            print(f"{route} - ₹{price}")

    def book_ticket(self):
        name = input("Passenger Name: ")
        age = int(input("Age: "))
        mobile = input("Mobile: ")

        self.show_routes()
        route = input("Select Route: ")

        if route not in self.routes:
            print("Invalid route")
            return

        if self.seats[route] >= 40:
            print(" Bus full")
            return

        self.ticket_id += 1
        self.seats[route] += 1

        self.tickets[self.ticket_id] = {
            "name": name,
            "route": route,
            "seat": self.seats[route],
            "price": self.routes[route]
        }

        print(f" Ticket Booked. Ticket ID: {self.ticket_id}")

    def view_ticket(self):
        tid = int(input("Enter Ticket ID: "))
        print(self.tickets.get(tid, "Ticket not found"))

    def cancel_ticket(self):
        tid = int(input("Ticket ID: "))
        if self.tickets.pop(tid, None):
            print("❌ Ticket cancelled")
        else:
            print("Ticket not found")

    def menu(self):
        while True:
            print("\n--- Bus Reservation ---")
            print("1. Show Routes")
            print("2. Book Ticket")
            print("3. View Ticket")
            print("4. Cancel Ticket")
            print("5. Exit")

            choice = input("Enter choice: ")

            if choice == "1":
                self.show_routes()
            elif choice == "2":
                self.book_ticket()
            elif choice == "3":
                self.view_ticket()
            elif choice == "4":
                self.cancel_ticket()
            elif choice == "5":
                break
```
