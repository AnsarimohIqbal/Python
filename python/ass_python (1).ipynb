16) Write a Python program to count the number of characters
(character frequency) in a string

```python
text = input("Enter a string: ")

freq = {}

for ch in text:
    if ch in freq:
        freq[ch] = freq[ch] + 1
    else:
        freq[ch] = 1

print("Character frequency:")
print(freq)
```

```python
def count_matching_strings(strings):
    count = 0

    for s in strings:
        if len(s) >= 2 and s[0] == s[-1]:
            count += 1

    return count


words = ['abc', 'xyz', 'aba', '1221', 'aa', 'a', 'x']
print(count_matching_strings(words))
```

```python
def is_sublist(main_list, sub_list):
    n = len(sub_list)

    for i in range(len(main_list) - n + 1):
        if main_list[i:i+n] == sub_list:
            return True
    return False



main_list = [1, 2, 3, 4, 5, 6]
sub_list = [3, 4, 5]

print(is_sublist(main_list, sub_list))
```

```python
def factorial(n):
    if n < 0:
        return "Factorial not defined for negative numbers"

    result = 1
    for i in range(1, n + 1):
        result *= i

    return result



print(factorial(5))
```

```python
def is_in_range(num, start, end):
    return start <= num <= end



print(is_in_range(5, 1, 10))   # True
print(is_in_range(15, 1, 10))  # False
```

```python
def is_perfect_number(n):
    if n <= 1:
        return False

    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i

    return total == n


```

```python
def is_palindrome(s):
    return s == s[::-1]



print(is_palindrome("madam"))   # True
print(is_palindrome("hello"))   # False
```

```python
words = ["python", "data", "programming", "ai", "machinelearning"]

max_len = 0
longest_words = []

for word in words:
    if len(word) > max_len:
        max_len = len(word)
        longest_words = [word]
    elif len(word) == max_len:
        longest_words.append(word)

print("Longest word(s):", longest_words)
print("Length:", max_len)
```
